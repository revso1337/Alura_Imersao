```python
#Bibliotecas para importação
import pandas as pd
import plotly.express as px
```

```python
#Pegando os valores da pasta "Principal" dentro da planilha e atribuindo o valor a uma variável
df_principal = pd.read_excel("/content/Aula 01.xlsx", sheet_name = "Principal")
```

```python
#Pegando os valores da pasta "Pesquisa ChatGPT" dentro da planilha e atribuindo o valor a uma variável
df_Pesquisa_ChatGPT = pd.read_excel("/content/Aula 01.xlsx", sheet_name = "Pesquisa ChatGPT")
```

```python
#Pegando os valores da pasta "Qntd de Acoes" dentro da planilha e atribuindo o valor a uma variável
df_Qntd_de_Acoes = pd.read_excel("/content/Aula 01.xlsx", sheet_name = "Qntd de Acoes")
```

```python
#Pegando os valores da pasta "Tickers" dentro da planilha e atribuindo o valor a uma variável
df_Tickers = pd.read_excel("/content/Aula 01.xlsx", sheet_name = "Tickers")
```

```python
#Pegando os valores da pasta "Analises" dentro da planilha e atribuindo o valor a uma variável
df_Analises = pd.read_excel("/content/Aula 01.xlsx", sheet_name = "Analises")
```

```python
#Copiando os valores das colunas da variável "df_principal" e substituindo para trabalharmos apenas com parte das colunas
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()

#Imprimindo(Mostrando) o conteúdo da variável mostrando somente as colunas desejadas
df_principal
```

```python
#Renomeiando as colunas para que não haja conflitos com os caracteres espeicias e/ou acentuações
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_final','Var. Dia (%)':'Var_dia_perct'}).copy()

#Imprimindo(Mostrando) o conteúdo da variável com a alteração do nome das colunas
df_principal
```

```python
#Pegando os valores da coluna "Var_dia_perct" dividindo por 100 e gerando uma uma coluna dentro da tabela com o resultado dessa divisão
df_principal['Var_prct'] = df_principal['Var_dia_perct'] / 100

#Gerando o valor inicial do ticker ao inicio do dia calculando o valor final e utilizando a divisão da porcertagem realizada ateriormente
# x = Valor_final / (Var_prct + 1)
df_principal['Valor_inicial'] = df_principal['Valor_final'] / (df_principal['Var_prct'] + 1)

#Imprimindo(Mostrando) o conteúdo da variável com a inclusão das colunas anteriormente calculadas
df_principal
```

```python
#Utilizando o comando .merge da biblioteca pandas para poder comparar dados de duas pastas da planilha
#Onde é comparado os valores de duas colunas "Ativo" de uma pasta com a coluna "Código" da outra pasta
#Com isso é copiado o conteúdo de uma pasta para outra
df_principal = df_principal.merge(df_Qntd_de_Acoes, left_on='Ativo', right_on='Código', how='left')

#Utilizado o comando .drop pra poder retirar a coluna "Código" que veio duplicada com os mesmos valores da outa coluna
df_principal = df_principal.drop(columns=['Código'])

#Imprimindo(Mostrando) o conteúdo da variável com a operação de comparação das colunas
df_principal
```

```python
#Realizando renomeiando a coluna "Qtde. Teórica" para não gerar possíveis conflitos
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'})

#Imprimindo(Mostrando) o conteúdo da variável com a alteração do nome da coluna
df_principal
```

```python
#Criando uma nova coluna para realizar a conta da variação de mercado da ação
df_principal['Var_em_Reais'] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtd_teorica']

#Imprimindo(Mostrando) o conteúdo da variável com a variação do valor das ações
df_principal
```

```python
#Alterando as informações numéricas para números decimais
pd.options.display.float_format = '{:.2f}' .format

#Imprimindo(Mostrando) o conteúdo da variável com a alteração dos valores para decimais
df_principal
```

```python
#Alterando o valor da coluna "Qtd_teorica" novamente para números inteiros
df_principal['Qtd_teorica'] = df_principal['Qtd_teorica'].astype(int)

#Imprimindo(Mostrando) o conteúdo da coluna "Qtd_teorica" com a alteração dos valores para inteiros
df_principal
```

```python
df_principal['Resultado'] = df_principal['Var_em_Reais'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.merge(df_Tickers, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_Pesquisa_ChatGPT, left_on='Nome', right_on='Nome Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome Empresa'])
df_principal = df_principal.rename(columns={'Idade (em anos)':'Idade'}).copy()
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['Idade'].apply(lambda x: 'Mais 100' if x > 100 else ('Menos 50' if x < 50 else '50 ate 100'))
df_principal
```

```python
# Calcular o máximo da coluna 'Variação'
maximo = df_principal['Var_em_Reais'].max()

# Calcular o mínimo da coluna 'Variação'
minimo = df_principal['Var_em_Reais'].min()

# Calcular a média da coluna 'Variação'
media = df_principal['Var_em_Reais'].mean()

# Calcular a média dos valores quando a variação é "Subiu"
media_subiu = df_principal.loc[df_principal['Resultado'] == 'Subiu', 'Var_em_Reais'].mean()

# Calcular a média dos valores quando a variação é "Desceu"
media_desceu = df_principal.loc[df_principal['Resultado'] == 'Desceu', 'Var_em_Reais'].mean()

print("Maior: R${:,.2f}".format(maximo))
print("Menor: R${:,.2f}".format(minimo))
print("Media: R${:,.2f}".format(media))
print("Média Subiu: R${:,.2f}".format(media_subiu))
print("Média Desceu: R${:,.2f}".format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento1 = df_principal_subiu.groupby('Segmento')['Var_em_Reais'].sum().reset_index()
df_analise_segmento1
```

```python
df_principal_desceu = df_principal[df_principal['Resultado'] == 'Desceu']
df_principal_desceu
```

```python
df_analise_segmento2 = df_principal_desceu.groupby('Segmento')['Var_em_Reais'].sum().reset_index()
df_analise_segmento2['Var_em_Reais'] = df_analise_segmento2['Var_em_Reais'].apply(lambda x: 'R$ {:.2f}'.format(x))
df_analise_segmento2
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Var_em_Reais'].sum().reset_index()
df_analise_saldo['Var_em_Reais_Formatado'] = df_analise_saldo['Var_em_Reais']
df_analise_saldo['Var_em_Reais'] = df_analise_saldo['Var_em_Reais'].apply(lambda x: 'R$ {:.2f}'.format(x))

df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Var_em_Reais_Formatado', text='Var_em_Reais', title='Variação Reais Por Resultado')
fig.update_xaxes(title_text='Resultados')
fig.update_yaxes(title_text='Valores')
fig.show()
```

```python
fig = px.pie(df_analise_segmento1, values='Var_em_Reais', names='Segmento', title='Variação em Reais por Segmento')
fig.update_traces(textinfo='label+percent', hovertemplate='Segmento: %{label}<br>Valor: R$ %{value:,.2f}', textposition='inside')
fig.show()
```

```python
df_idade_menos50 = df_principal[df_principal['Cat_Idade'] == 'Menos 50']
df_idade_50ate100 = df_principal[df_principal['Cat_Idade'] == '50 ate 100']
df_idade_mais100 = df_principal[df_principal['Cat_Idade'] == 'Mais 100']

# Concatenar os DataFrames
df_concatenado = pd.concat([df_idade_menos50, df_idade_50ate100, df_idade_mais100])

# Recalcular a análise por idade para o DataFrame concatenado
df_analise_idade_concatenado = df_concatenado.groupby('Cat_Idade').agg({'Var_em_Reais': ['sum', 'count']}).reset_index()
df_analise_idade_concatenado.columns = ['Cat_Idade', 'Var_em_Reais', 'Num_Empresas']
df_analise_idade_concatenado['Var_em_Reais_formatado'] = df_analise_idade_concatenado['Var_em_Reais']
df_analise_idade_concatenado['Var_em_Reais'] = df_analise_idade_concatenado['Var_em_Reais'].apply(lambda x: 'R$ {:.2f}'.format(x))

df_analise_idade_concatenado
```

```python
# Criar o gráfico de barras
fig = px.bar(df_analise_idade_concatenado, x='Cat_Idade', y='Var_em_Reais_formatado', text='Var_em_Reais', title='Soma das Variações em Reais por Faixa Etária')

# Adicionar rótulos aos eixos
fig.update_xaxes(title_text='Faixa Etária')
fig.update_yaxes(title_text='Soma das Variações')

# Exibir o gráfico
fig.show()
```

```python
# Criar o gráfico de barras
fig = px.bar(df_analise_idade_concatenado, x='Cat_Idade', y='Num_Empresas', text='Num_Empresas', title='Soma das Variações em Reais por Faixa Etária')

# Adicionar rótulos aos eixos
fig.update_xaxes(title_text='Faixa Etária')
fig.update_yaxes(title_text='Quantidade de empresas')

# Exibir o gráfico
fig.show()
```
